```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

pd.set_option('max.columns', None)
```

```python
df = pd.read_csv('Data/Power_Quality_data.csv')
df.sample(7)
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
df.sample(5)
```

```python
df.describe()
```

**All the features are continuous variables except the Target variable which is Discrete**

```python
df.isna().sum().sort_values(ascending = False) # no missing values present
```

```python
for col in df.columns:
    sns.histplot(data = df, x = col)
    plt.show()
```

- **The plots indicate that all values are centered around 0**
- **We will scale them later**

```python
df['output'].value_counts()  # the output variable seems balanced
```

```python
X = df.drop('output', axis = 1)
y = df['output']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
X_transf = scaler.fit_transform(X_train)
```

```python
X_transformed = pd.DataFrame(data = X_transf, columns = X.columns)
X_transformed.sample(7)
```

```python
X_test_tr = scaler.transform(X_test)
X_test_transformed = pd.DataFrame(data = X_test_tr, columns = X.columns)
X_test_transformed.sample(5)
```

```python
X_transformed.to_csv('Data/training_data.csv', index = False)
X_test_transformed.to_csv('Data/testing_data.csv', index = False)
y_train.to_csv('Data/y_train.csv', index = False)
y_test.to_csv('Data/y_test.csv', index = False)
```
